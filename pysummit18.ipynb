### Channel concepts

- A FIFO at-most-once queue
- Messegar producers and consumers
- Asynchronous: event-driven (Channels 1 "hides" async parts)
- ASGI: protocol specs (async version of WSGI)
    - Version 1 for Channel 1
    - Version 2 for Channel 2
- Protocol Server
    - Daphne for websockets
    - Defines channel
- Routing
    - map channels to consumers
    - similar to Django URLconf
- Consumers
    - read/write messages on the channel
    - similar to Django views (result is a message put on a channel)
- Frontend
    - Websockets means JS

### Demo

https://github.com/yakky/channeled-dashboard

Features (like Google Docs)
- Count active users
- Concurrency check (on documents)
- Browser notification

### Dive in
- Channel layers
    - Config (like Redis, routing)
- Routing
    - like URLconf - works on URLs
- Consumers
    - JsonWebsocketConsumer (fairly low-level, lot of control)
    - Counting users
    - Concurrency monitor
    - Browser notifications
        - streams are subchannels, used by the multiplexer

### Channel 2

- based on asyncio
- Python 3.5+
- mix of sync / async code
- protocol server and workers are on the same process as Django
- persistent consumers

### FAQ

- Performance: Workers can be scaled 
- Deployment: WSGI can be replaced by Daphne (or WSGI and map wecksockets via paths to Daphne)
